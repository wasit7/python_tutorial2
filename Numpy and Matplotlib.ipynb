```python
x=np.arange(start=0,stop=6*np.pi,step=0.1)
y=np.sin(x)
z=np.random.randn(x.shape[0])*0.2
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot(x,y,'r.')
plt.plot(x,z,'g.')
```

```python
a=y+z
plt.plot(x,a,'g.')
```

```python
plt.plot(x[a>0],a[a>0],'g.')
```
